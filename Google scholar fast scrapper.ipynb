```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import numpy as np
```

```python
path = 'C:\Program Files (x86)/chromedriver.exe'
```

```python
search_topic = "housing berlin"


driver = webdriver.Chrome(path)
driver.get('https://scholar.google.com/')
search = driver.find_element_by_name('q')
search.send_keys(search_topic)
search.send_keys(Keys.RETURN)

page = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.ID, 'gs_bdy')))


name = []
year = []
authors = []
publisher = []

n = 1
titles = page.find_elements_by_class_name('gs_rt')
articles = page.find_elements_by_class_name('gs_a')
for title, article in zip(titles, articles):
    name.append(title.text)
    year.append((article.text.replace('-',',').split(','))[-2])
    publisher.append((article.text.replace('-',',').split(','))[-1])
    authors.append(((article.text.replace('-',',').split(','))[:-2]))
    
elements =  page.find_elements_by_class_name('gs_nma')
links = [elem.get_attribute('href') for elem in elements]

n+=1

for link in links:
    try:
        driver.get(str(link))
        page = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.ID, 'gs_bdy')))

        titles = page.find_elements_by_class_name('gs_rt')
        articles = page.find_elements_by_class_name('gs_a')
        for title, article in zip(titles, articles):
            name.append(title.text)
            year.append((article.text.replace('-',',').split(','))[-2])
            publisher.append((article.text.replace('-',',').split(','))[-1])
            authors.append(((article.text.replace('-',',').split(','))[:-2]))
        
        n +=1
    except:
        pass
    
print(len(year))
print(len(name))
print(len(authors))
print(len(publisher))

results = pd.DataFrame(data = {'title':name, 'year':year, 'puplisher/source':publisher,'authors/journal':authors})
results.to_csv('search_results_'+str(search_topic).replace(' ','_')+'.csv')
results.head()
```
